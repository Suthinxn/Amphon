```python
import pandas as pd
import plotly.express as px
import numpy as np
data = pd.read_csv("air4thai_44t_2024-01-01_2024-02-29.csv")
data = pd.DataFrame(data=data)
del data['Unnamed: 0']
data['DATETIMEDATA'] = pd.to_datetime(data['DATETIMEDATA'])

fig = px.line(data, x="DATETIMEDATA", y="PM25", template="plotly_dark")
fig.show()
```

```python
data.info()
mean_pm = data['PM25'].mean()
data['PM25'].fillna(mean_pm, inplace = True)
data.info()
```

```python
data
```

```python
data['Hour'] = [i.hour for i in data['DATETIMEDATA']]
# data['Hour']
data['Day'] = [i.day for i in data['DATETIMEDATA']]
# data['Day']
data['Month'] = [i.month for i in data['DATETIMEDATA']]
# data['Month']
data['Year'] = [i.year for i in data['DATETIMEDATA']]
# data['Year']
data['Series'] = np.arange(1,len(data)+1)

data = data[['Series', 'Year', 'Month', 'Day', 'Hour', 'PM25', 'TEMP', 'WS', 'WD', 'RH', 'RAIN']]
data.head(48)
```

```python
result = data
```

```python
all_columns = result[result.columns[0]].count()
```

```python
test = result.sample(int(0.3*all_columns), random_state=20)
train = result.drop(test.index)
```

```python
from pycaret.regression import *
Setup_ = setup(train, target="PM25")
```

```python
best = compare_models(sort = 'MAE')
```

```python
evaluate_model(best)
```

```python
predict_model(best, data)
```

```python
predict_model(best)
```
